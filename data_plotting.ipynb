```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.array([6.52,6,5.55,5.26,4.72])
y = np.array([.6725, .585,.4975,.3725,.225])
x = x * math.pow(10, 14)
w, b = np.polyfit(x, y, 1)
planks_constant = 1.602176634 * math.pow(10, -19) * w
print('Planks constant = ', planks_constant)
fig, ax = plt.subplots()
ax.scatter(x,y)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Average Stopping Voltage (V)')
ax.plot(x, w*np.array(x) + b, color='red')
plt.text(x[1]+0.2*10**14, y[1], f"h = slope × e = {planks_constant:.2e} Js")
ax.set_xlim(left=4*10**14, right=7*10**14)
ax.set_ylim(bottom=0)
plt.grid()
plt.show()
```
